# Множини (Sets) у Python: Детальна документація та приклади

**Множина (set)** у Python — це неупорядкована колекція унікальних об'єктів. Вона працює так само, як математичні множини, і підтримує відповідні операції (перетин, об'єднання тощо).

Множини можна уявити як мішок, у який ви закидаєте речі: 
1. Ви не знаєте, в якому порядку вони там лежать.
2. Якщо ви покладете туди два однакових яблука, "мішок" залишить лише одне. Усі елементи абсолютно унікальні.

## 1. Базові принципи та створення
* **Швидкість:** Перевірка, чи є елемент у множині (`x in my_set`), працює миттєво ($O(1)$) завдяки хеш-таблицям під капотом.
* **Елементи:** Повинні бути незмінними (можна додати числа або рядки, але не можна додати список).

```python
# 1. Створення множини
fruits = {"яблуко", "банан", "апельсин"}
print("Початкова множина:", fruits)

# 2. Демонстрація унікальності (спроба додати те, що вже є)
fruits.add("яблуко")
fruits.add("яблуко")
print("Після додавання дублікатів:", fruits) # Нічого не зміниться!

# 3. ⚠️ ВАЖЛИВИЙ НЮАНС для новачків: Порожня множина
# Щоб створити порожню множину, треба використовувати set(), а не фігурні дужки!
empty_dict = {}     # Це створить порожній словник
empty_set = set()   # Це створить порожню множину

print("\nТип empty_dict:", type(empty_dict))
print("Тип empty_set:", type(empty_set))

# 4. Швидке видалення дублікатів зі списку
dirty_data = [1, 2, 2, 3, 3, 3, 4, 5, 5]
clean_data = list(set(dirty_data))
print("\nОчищений список від дублікатів:", clean_data)
```

## 2. Основні математичні операції

Множини підтримують потужні операції порівняння та об'єднання. 
* Ви можете використовувати спеціальні знаки-оператори (наприклад, `&`, `|`). Вони вимагають, щоб з обох боків були саме множини.
* Або методи (наприклад, `.intersection()`), які можуть приймати будь-що, наприклад звичайний список.

### Об'єднання (Union) `|`
Збирає всі елементи з обох множин до купи. Якщо елемент є і там, і там — він буде присутній лише один раз.

### Перетин (Intersection) `&`
Знаходить **лише спільні** елементи. Наприклад, пошук людей, які підписані на обох блогерів одночасно.

```python
# --- ПРИКЛАД: Соціальна мережа (Пошук друзів) ---

friends_user1 = {"Олена", "Іван", "Марія", "Петро"}
friends_user2 = {"Іван", "Марія", "Сергій", "Олег"}

# Перетин: Шукаємо спільних друзів
mutual_friends = friends_user1 & friends_user2
print("Спільні друзі:", mutual_friends)

# Об'єднання: Збираємо всіх унікальних друзів з двох списків
all_unique_friends = friends_user1 | friends_user2
print("Усі друзі (без повторень):", all_unique_friends)
```

### Різниця (Difference) `-`
Бере першу множину і "віднімає" (викидає) з неї всі елементи, які є у другій множині.

### Симетрична різниця (Symmetric Difference) `^`
Залишає лише ті елементи, які є унікальними для кожної з множин (викидає все спільне).

```python
# --- ПРИКЛАД: Перевірка виконання завдань ---

all_students = {"Максим", "Юлія", "Антон", "Світлана"}
submitted_homework = {"Максим", "Світлана"}

# Різниця: Хто зі студентів НЕ здав завдання?
# Логіка: Беремо всіх студентів і віднімаємо тих, хто вже здав.
lazy_students = all_students - submitted_homework
print("Не здали домашку:", lazy_students)

# Симетрична різниця: Хто має доступ ТІЛЬКИ до одного проекту?
project_A_team = {"Антон", "Максим"}
project_B_team = {"Антон", "Олена"}

# Відсікаємо Антона, бо він в обох проектах
exclusive_members = project_A_team ^ project_B_team
print("Працюють лише на одному проекті:", exclusive_members)
```

## 3. Зміна множин "на місці" та `frozenset`

Коли у вас величезні обсяги даних, створення нових множин після кожної операції забиває пам'ять. Для цього є методи, які змінюють існуючу множину без створення нової:
* `a.update(b)` або `a |= b` (додає)
* `a.intersection_update(b)` або `a &= b` (залишає спільне)
* `a.difference_update(b)` або `a -= b` (видаляє спільне)

### Що таке `frozenset`?
Звичайну множину `set` можна змінювати, тому її не можна зробити ключем у словнику `dict` або покласти всередину іншої множини. 
Для вирішення цього є **`frozenset`** (заморожена множина). Після створення вона назавжди стає незмінною.

```python
# Використання frozenset як ключа у словнику
# Уявімо кеш для розрахунку відстаней між містами. Відстань однакова в обидва боки.

routes_cache = {
    frozenset(["Київ", "Чернівці"]): 540,
    frozenset(["Львів", "Чернівці"]): 270
}

# Порядок міст у frozenset не має значення! Ми знайдемо маршрут незалежно від порядку.
search_key_1 = frozenset(["Чернівці", "Київ"])
search_key_2 = frozenset(["Львів", "Чернівці"])

print(f"Відстань Київ-Чернівці: {routes_cache[search_key_1]} км")
print(f"Відстань Львів-Чернівці: {routes_cache[search_key_2]} км")
```

## 4. Просунуті приклади використання (Для Інженерії Даних та ML)

Тут наведені реальні сценарії, де множини демонструють свою максимальну ефективність: оптимізація ETL-пайплайнів, підготовка даних для навчання нейромереж та обробка логів архітектури мікросервісів.

```python
import time

# --- ПРИКЛАД 1: Оптимізація швидкості пошуку в ETL-пайплайні ---
# Уявімо, що ми збираємо гідрологічні дані з датчиків для проекту GeoHydroAI.
# Нам потрібно перевірити, чи ми вже опрацьовували конкретний ID запису.

# Створюємо масив на 10 мільйонів записів
processed_ids_list = list(range(10_000_000)) 
processed_ids_set = set(processed_ids_list)
new_sensor_id = 9_999_999

# Пошук у списку (Python перевіряє кожен елемент по черзі)
start_time = time.time()
_ = new_sensor_id in processed_ids_list
list_time = time.time() - start_time

# Пошук у множині (Python миттєво знаходить адресу через хеш)
start_time = time.time()
_ = new_sensor_id in processed_ids_set
set_time = time.time() - start_time

print(f"Час пошуку в List: {list_time:.5f} сек")
print(f"Час пошуку в Set:  {set_time:.5f} сек")
# set_time може бути 0.0, тому додамо маленьке число для уникнення ділення на нуль
print(f"Множина швидша приблизно у {list_time / (set_time + 1e-9):,.0f} разів!\n")


# --- ПРИКЛАД 2: Валідація токенів під час підготовки LLM (SFT, DPO) ---
# Ми хочемо дізнатися, які специфічні слова містить новий датасет порівняно з базовим.

corpus_base = "штучний інтелект аналізує гідрологічні дані для прогнозування"
corpus_new = "гідрологічні дані важливі для навчання нейромереж"

vocab_base = set(corpus_base.split())
vocab_new = set(corpus_new.split())

# Які нові терміни з'явилися у новому датасеті? (Різниця)
new_unique_tokens = vocab_new - vocab_base
print(f"Нові токени для донавчання: {new_unique_tokens}")


# --- ПРИКЛАД 3: Оркестрація мікросервісів (vLLM / LiteLLM) ---
# Збираємо логи з різних воркерів. Треба отримати єдиний набір унікальних ID запитів, 
# щоб зрозуміти загальне навантаження на систему (Control Plane).

node_1_requests = ["req_001", "req_002", "req_003", "req_001"] # Воркер 1 має дублікати
node_2_requests = ["req_003", "req_004", "req_005"]            # Воркер 2

# Дедуплікація і злиття відбувається елегантно в один рядок
unique_cluster_requests = set(node_1_requests) | set(node_2_requests)
print(f"\nУнікальні запити з усього кластера: {unique_cluster_requests}")
```
