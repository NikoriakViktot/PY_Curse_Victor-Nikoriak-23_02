### Що виведе Python?

x, y = True, False

**Варіанти:** `True` або `False`

> ✏️ Запишіть прогноз, перш ніж запускати наступну клітинку.

````python
import ipywidgets as widgets
from IPython.display import display, clear_output

prediction_logic = ""

_pb2_w = widgets.Text(
    placeholder="Ваш прогноз",
    description="Прогноз:",
    layout=widgets.Layout(width="320px"))
_pb2_btn = widgets.Button(description="Зберегти прогноз", button_style="primary")
_pb2_out = widgets.Output()

def _on_pb2(b):
    global prediction_logic
    if not require_student():
        with _pb2_out:
            clear_output()
            print("⛔ Спочатку представтесь")
        return
    val = _pb2_w.value.strip()
    if not val:
        with _pb2_out:
            clear_output()
            print("⚠️ Введіть прогноз перед виконанням наступної клітинки")
        return
    prediction_logic = val
    _pb2_btn.disabled = True
    with _pb2_out:
        clear_output()
        print("✅ Прогноз збережено:", prediction_logic)

_pb2_btn.on_click(_on_pb2)
display(widgets.VBox([_pb2_w, _pb2_btn, _pb2_out]))
````

````python
import ipywidgets as widgets
from IPython.display import display, Markdown

_rev2_btn = widgets.Button(description="🐍 Результат та пояснення", button_style="info")
_rev2_out = widgets.Output()

def _on_rev2(b):
    if not require_student():
        with _rev2_out:
            print("⛔ Спочатку представтесь")
        return
    if not prediction_logic.strip():
        with _rev2_out:
            print("⚠️ Спочатку зробіть прогноз у клітинці 🔮")
        return
    b.disabled = True
    with _rev2_out:
        print(f"{STUDENT_NAME}, твій прогноз: {prediction_logic}")

        x, y = True, False
        result = not x and y or x

        print("\n🐍 Покроково:")
        print("x =", x)
        print("y =", y)
        print("not x        =", not x)
        print("not x and y  =", not x and y)
        print("... or x     =", not x and y or x)
        print("\nresult =", result)

        correct_answer = "True"
        if prediction_logic.strip() == correct_answer:
            print("\n✅ Правильний прогноз!")
        else:
            print("\n❌ Неправильно. Очікувалось:", correct_answer)

        display(Markdown("""
---

## ✅ Що сталося?

### Пріоритет: `not` > `and` > `or`

Вираз `not x and y or x` читається як `((not x) and y) or x`

| Крок | Вираз              | Значення |
|------|--------------------|----------|
| 1    | `not True`         | `False`  |
| 2    | `False and False`  | `False`  |
| 3    | `False or True`    | `True`   |

---

### Short-circuiting

```python
True or f()    # f() НЕ викликається
False and f()  # f() НЕ викликається
```

## 🔑 Правило

> Якщо сумніваєтесь — **ставте дужки**. Код стане зрозумілішим.
"""))

        try:
            submit_task(STUDENT_NAME, "lesson_04_exam", "Task_logic_operators")
        except Exception:
            print("\nℹ️ submit_task() не підключений")

_rev2_btn.on_click(_on_rev2)
display(widgets.VBox([_rev2_btn, _rev2_out]))
````

# ✅ Task 4 — if / elif / else

### Що виведе Python?

```python
score = 85
if score >= 60:
    print("Ви склали іспит!")
elif score >= 80:
    print("Ви отримали оцінку B!")

```

> ✏️ Запишіть **точну фразу** яку виведе Python, перш ніж запускати наступну клітинку.

````python
import ipywidgets as widgets
from IPython.display import display, clear_output

prediction_if = ""

_pb4_w = widgets.Text(
    placeholder="Ваш прогноз",
    description="Прогноз:",
    layout=widgets.Layout(width="320px"))
_pb4_btn = widgets.Button(description="Зберегти прогноз", button_style="primary")
_pb4_out = widgets.Output()

def _on_pb4(b):
    global prediction_if
    if not require_student():
        with _pb4_out:
            clear_output()
            print("⛔ Спочатку представтесь")
        return
    val = _pb4_w.value.strip()
    if not val:
        with _pb4_out:
            clear_output()
            print("⚠️ Введіть прогноз перед виконанням наступної клітинки")
        return
    prediction_if = val
    _pb4_btn.disabled = True
    with _pb4_out:
        clear_output()
        print("✅ Прогноз збережено:", prediction_if)

_pb4_btn.on_click(_on_pb4)
display(widgets.VBox([_pb4_w, _pb4_btn, _pb4_out]))
````

````python
import ipywidgets as widgets
from IPython.display import display, Markdown
import io, sys

_rev4_btn = widgets.Button(description="🐍 Результат та пояснення", button_style="info")
_rev4_out = widgets.Output()

def _on_rev4(b):
    if not require_student():
        with _rev4_out:
            print("⛔ Спочатку представтесь")
        return
    if not prediction_if.strip():
        with _rev4_out:
            print("⚠️ Спочатку зробіть прогноз у клітинці 🔮")
        return
    b.disabled = True
    with _rev4_out:
        print(f"{STUDENT_NAME}, твій прогноз: {prediction_if}")

        # Capture stdout without losing the Output widget stream
        _prev_stdout = sys.stdout
        buf = io.StringIO()
        sys.stdout = buf

        score = 85
        if score >= 60:
            print("Ви склали іспит!")
        elif score >= 80:
            print("Ви отримали оцінку B!")
        else:
            print("Іспит не складено.")

        sys.stdout = _prev_stdout
        actual_output = buf.getvalue().strip()

        print("\n🐍 Python повернув:", actual_output)

        correct_answer = "Ви склали іспит!"
        if prediction_if.strip() == correct_answer:
            print("\n✅ Правильний прогноз!")
        else:
            print("\n❌ Неправильно. Очікувалось:", correct_answer)

        display(Markdown("""
---

## ✅ Що сталося?

### Перевірка зверху вниз

```python
score = 85
if score >= 60:    # 85 >= 60 → True ✅ Виконується!
    print("Ви склали іспит!")
elif score >= 80:  # ← ніколи не перевіряється
    print("Ви отримали оцінку B!")
```

Python знаходить **першу** умову `True`, виконує блок і **ігнорує** решту.

```python
# ✅ Правильний порядок — від строгішого до загального
if score >= 90:   print("Відмінно")
elif score >= 80: print("Добре")
elif score >= 60: print("Склав")
else:             print("Не склав")
```

## 🔑 Головне правило

> У `if-elif-else` виконується **лише перший** блок з True-умовою.  
> Розміщуйте умови від **конкретних** до **загальних**.
"""))

        try:
            submit_task(STUDENT_NAME, "lesson_04_exam", "Task_if_elif")
        except Exception:
            print("\nℹ️ submit_task() не підключений")

_rev4_btn.on_click(_on_rev4)
display(widgets.VBox([_rev4_btn, _rev4_out]))
````

# ✅ Task 5 — break та continue

### Що виведе Python?

```python
for num in range(1, 6):
    if num == 3:
        continue
    if num == 5:
        break
    print(num)
```

> ✏️ Введіть числа через пробіл (наприклад: `1 2 4`).

````python
import ipywidgets as widgets
from IPython.display import display, clear_output

prediction_loop = ""

_pb5_w = widgets.Text(
    placeholder="Ваш прогноз",
    description="Прогноз:",
    layout=widgets.Layout(width="320px"))
_pb5_btn = widgets.Button(description="Зберегти прогноз", button_style="primary")
_pb5_out = widgets.Output()

def _on_pb5(b):
    global prediction_loop
    if not require_student():
        with _pb5_out:
            clear_output()
            print("⛔ Спочатку представтесь")
        return
    val = _pb5_w.value.strip()
    if not val:
        with _pb5_out:
            clear_output()
            print("⚠️ Введіть прогноз перед виконанням наступної клітинки")
        return
    prediction_loop = val
    _pb5_btn.disabled = True
    with _pb5_out:
        clear_output()
        print("✅ Прогноз збережено:", prediction_loop)

_pb5_btn.on_click(_on_pb5)
display(widgets.VBox([_pb5_w, _pb5_btn, _pb5_out]))
````

````python
import ipywidgets as widgets
from IPython.display import display, Markdown

_rev5_btn = widgets.Button(description="🐍 Результат та пояснення", button_style="info")
_rev5_out = widgets.Output()

def _on_rev5(b):
    if not require_student():
        with _rev5_out:
            print("⛔ Спочатку представтесь")
        return
    if not prediction_loop.strip():
        with _rev5_out:
            print("⚠️ Спочатку зробіть прогноз у клітинці 🔮")
        return
    b.disabled = True
    with _rev5_out:
        print(f"{STUDENT_NAME}, твій прогноз: {prediction_loop}")

        print("\n🐍 Python виводить (покроково):")
        for num in range(1, 6):
            if num == 3:
                print(f"  num={num} → continue (пропускаємо)")
                continue
            if num == 5:
                print(f"  num={num} → break (зупиняємось)")
                break
            print(f"  num={num} → print({num})")

        print("\n📌 Фактичний вивід:")
        for num in range(1, 6):
            if num == 3:
                continue
            if num == 5:
                break
            print(num)

        correct_answer = "1 2 4"
        if prediction_loop.strip() == correct_answer:
            print("\n✅ Правильний прогноз!")
        else:
            print("\n❌ Неправильно. Очікувалось:", correct_answer)

        display(Markdown("""
---

## ✅ Що сталося?

| Ітерація | num | Дія          | Виводить? |
|----------|-----|--------------|----------|
| 1        | 1   | звичайна     | ✅ `1`   |
| 2        | 2   | звичайна     | ✅ `2`   |
| 3        | 3   | `continue`   | ❌ пропуск |
| 4        | 4   | звичайна     | ✅ `4`   |
| 5        | 5   | `break`      | ❌ вихід |

**Результат:** `1 2 4`

```python
# break — повністю зупиняє цикл
# continue — пропускає лише цю ітерацію
```

## 🔑 Головне правило

> `break` = вийти з циклу назавжди.  
> `continue` = перейти до наступної ітерації.
"""))

        try:
            submit_task(STUDENT_NAME, "lesson_04_exam", "Task_break_continue")
        except Exception:
            print("\nℹ️ submit_task() не підключений")

_rev5_btn.on_click(_on_rev5)
display(widgets.VBox([_rev5_btn, _rev5_out]))
````

````python
# ── EXAM — питання з Google Sheets ────────────────────────────────────────────
import uuid, time
import ipywidgets as widgets
from IPython.display import display, clear_output

_API  = "https://script.google.com/macros/s/AKfycbyoTGSLxFNEFkhwfYLDUrOwGrIHKWsIQ1SKOPu8AvNlSUYMlm9h7yK0kGYWD45a1Loa/exec"
_LSID = "lesson_04_exam"

_LEVEL_ORDER  = ["bronze", "silver", "gold", "platinum"]
_LEVEL_LABELS = {
    "bronze":   "🟢 Bronze",
    "silver":   "🟡 Silver",
    "gold":     "🟠 Gold",
    "platinum": "🔴 Platinum",
}

_exam_start_btn = widgets.Button(
    description="▶ Завантажити іспит",
    button_style="success",
    layout=widgets.Layout(width="220px"))
_exam_out = widgets.Output()

def _on_exam_start(b):
    if not require_student():
        with _exam_out:
            print("⛔ Спочатку представтесь")
        return
    b.disabled = True
    b.description = "⏳ Завантаження..."

    with _exam_out:
        print("⏳ Підключення до сервера...")
        try:
            _sess = requests.get(_API, params={"name": STUDENT_NAME, "lesson_id": _LSID}).json()
        except Exception as ex:
            print("❌ Помилка з'єднання:", ex)
            b.disabled = False
            b.description = "▶ Завантажити іспит"
            return

        if not _sess.get("ok"):
            print("❌ Помилка сесії:", _sess.get("error"))
            return

        _token = _sess["token"]
        _qr = requests.get(_API, params={"action": "questions", "token": _token}).json()
        if not _qr.get("ok"):
            print("❌ Питання не завантажено:", _qr.get("error"))
            return

        _questions = _qr.get("questions", [])
        print(f"✅ Завантажено {len(_questions)} питань  |  {STUDENT_NAME}, починайте!\n")

        def _submit_exam(task_id, answer):
            try:
                r = requests.post(_API, json={
                    "token":     _token,
                    "task_id":   task_id,
                    "result":    {"answer": answer},
                    "nonce":     str(uuid.uuid4()),
                    "client_ts": int(time.time() * 1000),
                })
                return r.json()
            except Exception as ex:
                return {"ok": False, "error": str(ex)}

        def _make_question(q):
            tid   = q.get("id", "")
            qtext = q.get("question", "")
            opts  = [str(o) for o in q.get("options", []) if str(o).strip()]

            q_w   = widgets.HTML(
                f'<div style="font-size:14px;font-weight:600;margin-bottom:6px">{qtext}</div>')
            opt_w = widgets.RadioButtons(options=opts,
                                          layout=widgets.Layout(width="max-content"))
            btn_w = widgets.Button(description="Перевірити", button_style="primary",
                                   layout=widgets.Layout(margin="4px 0 0 0"))
            out_w = widgets.Output()
            _done = [False]

            def _click(b):
                if _done[0]:
                    return
                _done[0]       = True
                btn_w.disabled = True
                ans = opt_w.value
                with out_w:
                    print("⏳ Перевірка...")
                resp = _submit_exam(tid, ans)
                prog = resp.get("progress", {})
                ok   = prog.get("correct", False)
                with out_w:
                    clear_output()
                    print("✅ Правильно!" if ok else "❌ Неправильно")
                    pct   = prog.get("score_pct", "—")
                    cnt   = prog.get("correct_count", "—")
                    ans_n = prog.get("answered", "—")
                    print(f"📊 Загалом: {cnt}/{ans_n} ({pct}%)")

            btn_w.on_click(_click)
            return widgets.VBox(
                [q_w, opt_w, btn_w, out_w],
                layout=widgets.Layout(
                    margin="0 0 10px 0", padding="12px 16px",
                    border="1px solid #ddd", border_radius="8px"))

        _by_level = {}
        for _q in _questions:
            _lvl = _q.get("level", "").lower().strip()
            _by_level.setdefault(_lvl, []).append(_q)

        for _lvl in _LEVEL_ORDER:
            _qs = _by_level.get(_lvl, [])
            if not _qs:
                continue
            _label = _LEVEL_LABELS.get(_lvl, _lvl.title())
            display(widgets.HTML(f'<h2 style="margin-top:20px;margin-bottom:8px">{_label}</h2>'))
            for _q in _qs:
                display(_make_question(_q))

_exam_start_btn.on_click(_on_exam_start)
display(widgets.VBox([_exam_start_btn, _exam_out]))
````
